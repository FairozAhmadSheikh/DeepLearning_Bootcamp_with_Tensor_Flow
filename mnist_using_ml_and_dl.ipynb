```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

# Load and Prepare Data

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# Normalize the data
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

```python
print(f"Training data shape: {X_train.shape}")
print(f"Number of classes: {len(tf.unique(y_train).y.numpy())}")
```

Baseline Model (Simple MLP/Shallow NN)

```python
baseline_model = Sequential([
    # Input layer
    Flatten(input_shape=(28, 28)), 
    # Single Hidden Layer 
    Dense(128, activation='relu'), 
    # Output layer: 10 neurons for 10 classes (digits 0-9)

    Dense(10, activation='softmax')
])
```

```python
# Compile the baseline model
baseline_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
print("Train Baseline Model")
baseline_history = baseline_model.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test, y_test),
    verbose=0 # set to 1 for progress bar
)
```

```python
_, baseline_acc = baseline_model.evaluate(X_test, y_test, verbose=0)
print(f"Baseline Model Test Accuracy: {baseline_acc:.4f}")
```
